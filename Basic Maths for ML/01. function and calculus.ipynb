```python
# find Slope of f(x) at x = 2 

f = x ** 3
d_dx = diff(f)
print(d_dx)
print(d_dx.subs(x,2))
```

```python
x,y = symbols('x y')
f = 2*x**3 + 3*y**3

d_dx = diff(f,x)

print("partial derivate of f(x,y) w.r.t x :: ", d_dx)

d_dy = diff(f,y)

print("partial derivate of f(x,y) w.r.t x :: ", d_dy)
```

```python
#Chain rule
x,y = symbols('x y')

_y = x**2 + 1
z = y**3 - 2

dy_dx = diff(_y, x)
dz_dy = diff(z, y)

chain_rule_ans_dz_dx = dy_dx * dz_dy

print("Chain rule derivate of z w.r.t x :: ", chain_rule_ans_dz_dx )

# substitution 
without_chain_rule_dz_dx = diff(z.subs(y, _y))

print("Without Chain rule derivate of z w.r.t x :: ", without_chain_rule_dz_dx )
```

## Integration

$$
\int_{0}^{x} x^2 \,dx
$$

```python
x = symbols('x')

f = x**2

result = integrate(f, (x,0,x))
print(result)
```

$$
\int_{1}^{2} 2*x^3 \,dx
$$

```python
x = symbols('x')

f = 2*x**3
result = integrate(f, (x,1,2))
print(result)
```

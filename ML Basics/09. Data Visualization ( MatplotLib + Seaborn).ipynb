# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Rooms':[1,1,1,2,2,3,3,4,4,4,5],'Price':[100,115,120,140,150,170,180,200,210,230,300]}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['Rooms'],df['Price']);      #  include ';' at last to avoid extra text print
```

```python
plt.scatter(df['Rooms'],df['Price']);
```

```python
# change color of line and type , include marker 
plt.plot(df['Rooms'],df['Price'], color = 'red', linestyle='--', marker = 'o', markersize =10);
```

```python
#Include title, label and set limit on axis value
plt.scatter(df['Rooms'],df['Price'], color= 'green')
plt.xlim(0,6)
plt.ylim(0,310)

plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.title("SCATTER PLOT");
```

# SeaBorn

```python
import seaborn as sns
```

```python
data = pd.read_csv('Data/Seaborn_heart_data.csv')
```

```python
data.head(5)
```

```python
sns.displot(data['age'], kde = True);     #distplot() is deprecated in future so use displot with kde as TRUE.
```

```python
# change bins number
sns.displot(data['age'], kde = True, bins= 30, color='red');
```

```python
# count distinct values of 'x' in dataset 
sns.countplot(x = 'sex', data=data)
```

```python
# hue feature in countplot
sns.countplot(x = 'cp', data=data, hue= 'sex');
```

# Box Plot

Box plots show the distributions across different categories.

<img src='Data/boxplot.png' style="max-width:50%;"></img>


```python
plt.figure(figsize= (15,9))
sns.boxplot(x='target',y='thalach',data=data,hue= 'sex')
```

```python
#scatterplot
plt.figure( figsize =(12,6))
sns.scatterplot(x='chol',y='trestbps',data=data,hue='sex');
```

```python
# cover 4 variable as one plot
plt.figure( figsize =(15,9))
sns.scatterplot(x='chol',y='trestbps',data=data,hue='sex',size='age', sizes=(1,150))
```

### Pairplots

Pairplots perform scatterplots and histograms for every single column in your data set. This means it could be a huge plot for large datasets! Use with caution, as it could take a long time for large datasets and the figures could be too small!


```python
from sklearn import datasets
iris = datasets.load_iris()

df = pd.DataFrame(data = iris['data'], columns = ['sepal_length','sepal_width','petal_length','petal_width'])
df['species'] = iris['target']
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue= 'species', palette='Dark2');
```
